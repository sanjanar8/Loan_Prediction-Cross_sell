```python
from sklearn.preprocessing import LabelEncoder
cat_columns = ['PREFERRED_CHANNEL','LAST_CAMPAIGN_RESPONSE']

data_encoded = data.copy()
encoders = {}

for col in cat_columns:
    le = LabelEncoder()
    data_encoded[col] = le.fit_transform(data_encoded[col].astype(str))
    encoders[col] = le
```

```python
data_encoded.head()
```

```python
data_encoded.info()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.callbacks import EarlyStopping
```

```python
X = data_encoded.drop(columns='CURRENT_LOAN_OWNED')

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train,X_val = train_test_split(
    X_scaled,
    test_size = 0.2,
    random_state = 42
)
```

```python
X_scaled[0]
```

```python
INPUT_DIM = X_scaled.shape[1]
EMBEDDING_DIM = 3

input_layer = Input(shape=(INPUT_DIM,))

##Encoder
x = Dense(64, activation="relu")(input_layer)
x = Dense(32, activation="relu")(x)

embedding = Dense(
    EMBEDDING_DIM,
    activation="linear",
    name = "embedding"
)(x)

##Decoder
x = Dense(32, activation="relu")(embedding)
x = Dense(64, activation="relu")(x)
output_layer = Dense(INPUT_DIM, activation='linear')(x)

autoencoder = Model(inputs=input_layer, outputs = output_layer)
encoder = Model(inputs=input_layer,outputs = embedding)
```

```python
autoencoder.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss="mse"
)

early_stop = EarlyStopping(
    monitor = "val_loss",
    patience = 5,
    restore_best_weights = True
)
```

```python
history = autoencoder.fit(
    X_train,
    X_train,
    validation_data = (X_val,X_val),
    epochs = 100,
    batch_size = 256,
    callbacks = [early_stop],
    verbose = 1
)
```

```python
customer_embeddings = encoder.predict(X_scaled)

embeddings_df = pd.DataFrame(
    customer_embeddings,
    index = data["SK_ID_CURR"],
    columns = [f"emb_{i}" for i in range(EMBEDDING_DIM)]
)
```

```python
embeddings_df.head()
```

```python
from sklearn.neighbors import NearestNeighbors

knn = NearestNeighbors(
    n_neighbors=5,
    metric="cosine"
)

knn.fit(embeddings_df.values)
```

```python
def X_sell_product(
    new_cust:dict,
    top_k:int = 5,
    top_n_products:int = 3
):
    new_df = pd.DataFrame([new_cust])
    new_scaled = scaler.transform(new_df)

    new_emb = encoder.predict(new_scaled)

    dist,indices = knn.kneighbors(new_emb,n_neighbors=top_k)

    neighbours = data.iloc[indices[0]]
    rec_df = pd.DataFrame(
        {
            "product": neighbours["CURRENT_LOAN_OWNED"].values,
            "distance": dist[0]
        }
    )
    rec_df["weight"] = 1 / (rec_df["distance"] + 1e-6)

    return (
        rec_df.groupby("product")["weight"].sum().sort_values(ascending=False).head(top_n_products)
    )
```
